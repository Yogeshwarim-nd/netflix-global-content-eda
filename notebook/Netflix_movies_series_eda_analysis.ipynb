## 🎬 Insights from Movie and TV Show Trends (2016–2018)

- Movies were the **primary contributors to the sharp increase in total releases between 2016 and 2018**.

- During this period, the **number of movies released was more than double the count of TV shows**, indicating a strong expansion in movie production and availability on streaming platforms.

- This surge can largely be attributed to the **global expansion of streaming services such as Netflix and Amazon Prime Video**, which invested heavily in **licensing and producing movies** to quickly expand their content libraries.

- The **inclusion of international and regional films**, along with improvements in **digital production and distribution**, further accelerated this growth and made it easier for **independent creators** to release their films online.

- **TV shows also showed a steady upward trend over the years**, reflecting the growing demand for long-form and episodic content.

- By **2021, the number of TV shows released surpassed the number of movies**, highlighting a shift in content strategy toward series-based entertainment on streaming platforms.

```python
# --- Filter Movies and TV Shows ---
movie_df = df[df['type'] == 'Movie'].copy()
tv_df = df[df['type'] == 'TV Show'].copy()
# --- Convert 'date_added' to datetime ---
movie_df['date_added'] = pd.to_datetime(movie_df['date_added'], errors='coerce')
tv_df['date_added'] = pd.to_datetime(tv_df['date_added'], errors='coerce')
# --- Extract month names ---
movie_df['month'] = movie_df['date_added'].dt.month_name()
tv_df['month'] = tv_df['date_added'].dt.month_name()
# --- Define month order ---
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
'July', 'August', 'September', 'October', 'November', 'December']
# --- Count number of Movies and TV Shows added per month ---
movie_month_wise = movie_df['month'].value_counts().reindex(month_order).fillna(0)
tv_month_wise = tv_df['month'].value_counts().reindex(month_order).fillna(0)
# --- Identify peak months ---
best_movie_month = movie_month_wise.idxmax()
best_tv_month = tv_month_wise.idxmax()

# --- Create color palettes ---
movie_colors = ['tomato' if month == best_movie_month else 'skyblue' for month in movie_month_wise.index]
tv_colors = ['orange' if month == best_tv_month else 'lightgreen' for month in tv_month_wise.index]
# --- Plot Movies by Month ---
plt.figure(figsize=(12,5))
plt.bar(movie_month_wise.index, movie_month_wise.values, color=movie_colors)
plt.xticks(rotation=45)
plt.title('Number of Movies Added by Month', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.text(
month_order.index(best_movie_month),
movie_month_wise.max(),
f'Peak: {best_movie_month}',
ha='center', va='bottom',
fontsize=12, color='darkred', fontweight='bold'
)
print(movie_month_wise)               
plt.show()
             
# --- Plot TV Shows by Month ---
plt.figure(figsize=(12,5))
plt.bar(tv_month_wise.index, tv_month_wise.values, color=tv_colors)
plt.xticks(rotation=45)
plt.title('Number of TV Shows Added by Month', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of TV Shows', fontsize=12)
plt.text(
month_order.index(best_tv_month),
tv_month_wise.max(),
f'Peak: {best_tv_month}',
ha='center', va='bottom',
fontsize=12, color='darkred', fontweight='bold'
)
print(tv_month_wise)
plt.show()
```

## 📅 Insights: Ideal Time to Launch Content

From the analysis of monthly content additions, we observe clear patterns in when Netflix tends to release new content.

- The **highest number of TV shows** were added to the platform in **December**, indicating a strong year-end release trend.
- **Movies reached their peak additions in July**, suggesting that mid-year is a preferred time for movie releases.
- These patterns indicate that **July and December are strategic periods for launching new content**, as streaming platforms increase releases during these months to maximize viewer engagement.

### Possible Reasons for This Trend

**🌞 July: Mid-Year Engagement Boost**

- Viewers often have **more leisure time during summer holidays**, leading to increased streaming activity.
- This period is particularly suitable for **movie releases**, as audiences tend to prefer shorter, standalone content during vacations.

**🎄 December: Year-End Holiday Surge**

- December coincides with **holiday vacations and festive breaks**, when audiences have more free time.
- This creates an ideal opportunity for viewers to **binge-watch TV shows and series**.

**🗓 Content Scheduling Strategy**

- Streaming platforms like Netflix may strategically **align major releases with mid-year and year-end marketing campaigns** to maximize visibility and audience reach.

**⏰ Higher Audience Availability**

- Seasonal breaks during **July and December** often lead to **longer viewing hours**, resulting in higher overall engagement.

### 📊 Business Recommendation

Launching **movies in July** and **TV shows in December** could help streaming platforms maximize **audience reach, viewer engagement, and marketing impact**.

```python
df_exploded = df.explode('country')
# Count number of movies and TV shows per country
country_counts = df_exploded.groupby(['country', 'type'],observed=False)['title'].count().reset_index()
country_counts.rename(columns={'title':'count'}, inplace=True)
# Pivot to have separate columns for Movies and TV Shows
pivot = country_counts.pivot(index='country', columns='type', values='count').fillna(0)
# Select top 10 countries by total content
pivot['total'] = pivot.sum(axis=1)
top10 = pivot.sort_values('total', ascending=False).head(10)
# Plot side-by-side bar chart
top10[['Movie','TV Show']].plot(kind='bar', figsize=(10,3))
plt.title('Top 10 Countries: Movies vs TV Shows on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.tight_layout()
print(top10)
plt.show()
```

```python
# Explode 'country' and 'listed_in' columns
df_exploded = df.explode('country').explode('listed_in')

# Get top 10 countries by total content
country_counts = df_exploded.groupby('country')['title'].count().reset_index()

top10_countries = country_counts.sort_values('title', ascending=False).head(10)['country'].tolist()

# Filter for top 10 countries
df_top10 = df_exploded[df_exploded['country'].isin(top10_countries)]

# -------------------------
# Step 2: Separate Movies and TV Shows
# -------------------------
movies_df = df_top10[df_top10['type'] == 'Movie']
tvshows_df = df_top10[df_top10['type'] == 'TV Show']

# -------------------------
# Step 3: Count genres per country
# -------------------------
movies_counts = movies_df.groupby(['country','listed_in'])['title'].count().reset_index()
movies_counts.rename(columns={'title':'count'}, inplace=True)

tvshows_counts = tvshows_df.groupby(['country','listed_in'])['title'].count().reset_index()
tvshows_counts.rename(columns={'title':'count'}, inplace=True)

# -------------------------
# Step 4: Plot top 5 genres for Movies
# -------------------------
sns.set(style="whitegrid")

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(18,25))
axes = axes.flatten()

for i, country in enumerate(top10_countries):
    
    ax = axes[i]

    data = movies_counts[movies_counts['country'] == country] \
            .sort_values('count', ascending=False).head(5)

    sns.barplot(data=data, y='listed_in', x='count', color='#1f77b4', ax=ax)

    # Add numbers on bars
    for p in ax.patches:
        width = p.get_width()
        ax.annotate(f'{int(width)}',
                    (width + 0.5, p.get_y() + p.get_height()/2),
                    va='center')

    ax.set_title(f'{country} - Top 5 Movie Genres', fontsize=14)
    ax.set_xlabel('Number of Titles', fontsize=12)
    ax.set_ylabel('Genre', fontsize=12)

plt.tight_layout()
plt.show()
```

## 🎬 Insight: Movie Production by Genre in Top Countries

From the analysis of **movie genres across the top producing countries on Netflix**, several regional patterns can be observed:

- **India, France, Japan, South Korea, Spain, Germany, and Mexico** show a higher production of **International Movies** compared to other genres.
- In the **United States** and **United Kingdom**, **Drama movies** dominate production.
- In **Canada**, **Comedy movies** have the highest production among all genres.

### 🌍 Key Observation
These patterns highlight how **different countries focus on specific movie genres** based on **audience preferences, cultural storytelling styles, and regional entertainment trends**.

Understanding these trends helps Netflix **tailor content strategies and recommendations for viewers in different markets worldwide**.
